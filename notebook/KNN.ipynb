```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split as tspl
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
#load data
data = pd.read_csv("../data/data.csv")
data.head()
```

```python
#delete first column
data.index = data['id']
data.drop(["id", "Unnamed: 32"], axis = 1, inplace = True)
data.head()
```

```python
diagnosis = data.diagnosis
data.drop(['diagnosis'], axis = 1, inplace = True)
```

```python
data_train, data_test, diag_train, diag_test = tspl(data, diagnosis, test_size = 0.33)
```

```python
for i in range(1,20):
    knn = KNN(n_neighbors = i)
    knn.fit(data_train, diag_train)
    score = knn.score(data_test, diag_test)
    print(str(i) + "  score = " + str(score))
```

seems that 5 neighbors give the highest score

```python
knn = KNN(n_neighbors = 5)
knn.fit(data_train, diag_train)
```

```python
prediction_probability = knn.predict_proba(data_test)
predictions = knn.predict(data_test)

from sklearn.metrics import confusion_matrix as cfm
cfm(predictions, diag_test)
```
